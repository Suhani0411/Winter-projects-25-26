```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/housing_prices - housing_prices.csv')
df.head()
```

```python
x = df.iloc[:, 0].to_numpy(dtype=float)   # area (square feet)
y = df.iloc[:, 1].to_numpy(dtype=float)   # price
```

```python
n=len(x)
sum_xy=sum(x*y)
sum_xsq=sum(x*x)
sum_ysq=sum(y*y)
sum_x=sum(x)
sum_y=sum(y)
```

```python
m=(n*(sum_xy)-(sum_x*sum_y))/(n*(sum_xsq)-(sum_x**2))
```

```python
b=sum_y/n - m*(sum_x/n)
```

```python
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
```

```python
def prediction_value(area):
  return m*area + b
```

```python
area_given = 2500
predicted_price = prediction_value(area_given)

print(f"Predicted price for {area_given}: {predicted_price:.2f}")
```

Plotting line of best fit

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y, label="Data", color="blue", alpha=0.6)

x1 = np.linspace(min(x), max(x), 200)
y1 = m*x1 + b
plt.plot(x1, y1, color="red", label=f"Best Fit Line")

# Predicted point
plt.scatter([area_given], [predicted_price], color="green",label="Prediction (2500)")

plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Housing Prices")
plt.legend()
plt.show()
```
